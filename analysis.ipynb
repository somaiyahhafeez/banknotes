```python
import pandas as pd
df= pd.read_csv('banknotesData by Pudding.csv')
df
```

```python
df = df.drop_duplicates('name')
```

```python
df['gender'].value_counts()
```

```python
dfgender_distribution = df.groupby(['country', 'gender']).size().unstack(fill_value=0)

# Sort by the 'F' column in descending order (more women first)
gender_distribution = gender_distribution.sort_values(by='F', ascending=False)

# Get the top 5 countries with the highest number of women
top_5_countries = gender_distribution.head(5)

print(top_5_countries)
```

```python
gender_distribution = df.groupby(['country', 'gender']).size().unstack(fill_value=0)

gender_distribution = gender_distribution.sort_values(by='F', ascending=False)

top_5_countries = gender_distribution.head(5)

print(top_5_countries)

```
